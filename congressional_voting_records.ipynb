```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data", 
                 names = ['party', 'infants', 'water', 'budget', 'physician','salvador', 'religious', 'satellite', 'aid', 'missile',
                          'immigration', 'synfuels', 'education', 'superfund', 'crime', 'duty_free_exports','eaa_rsa'])
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.info()
```

# **Tiền xử lý dữ liệu**

```python
df.replace(('y', 'n', '?'), (1, 0, np.NaN), inplace=True)
df.head()
```

**Điền các giá trị còn thiếu (NaN)**
bằng bfill (lấy các giá trị kế tiếp) và fffill (lấy các giá trị trước đó)







```python
df.fillna(method='bfill', inplace=True)
df.tail()
```

```python
df.fillna(method='ffill', inplace=True)
df.tail()
```

**Chuyển các feature về các giá trị 0/1**

```python
df.party.unique()
```

```python
df.party.replace(('republican', 'democrat'), (1, 0), inplace=True)
df.info()
```

```python
for col in df.columns:
  df[col] = pd.to_numeric(df[col]).astype(int)

df.info()
```

```python
df.head()
```

# **Data Modelling**

```python
y = df.party.values #features
X = df.drop('party', axis=1) # target
```

```python
X.shape
```

```python
y.shape
```

**Chia tập dữ liệu**
(stratify: phân chia phân tầng)

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25, random_state = 42, stratify=y)
```

```python
Xtrain.shape, Xtest.shape, ytrain.shape, ytest.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = np.arange(1, 9)
trainAccuracy = np.empty(len(neighbors))
testAccuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
  kNN = KNeighborsClassifier(n_neighbors=k)
  kNN.fit(Xtrain, ytrain)

  # Tính toán hệ số xác định cho tập train và tập test
  trainAccuracy[i] = kNN.score(Xtrain, ytrain)
  testAccuracy[i] = kNN.score(Xtest, ytest)

# score_acc = []
# for k in neighbors:
#   kNN = KNeighborsClassifier(n_neighbors=k)
#   kNN.fit(Xtrain, ytrain)
#   y_pred = kNN.predict(Xtest)
#   score_acc.append(metrics.accuracy_score(ytest, y_pred))
# score_acc
```

```python
plt.title('k-NN')
plt.plot(neighbors, trainAccuracy, label='Training Accuracy')
plt.plot(neighbors, testAccuracy, label='Testing Accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
# Chọn k = 3
kNN = KNeighborsClassifier(n_neighbors=3)
model = kNN.fit(Xtrain, ytrain)

y_pred = model.predict(Xtest)
```
