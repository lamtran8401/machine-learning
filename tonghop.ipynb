```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data"
names = ['party', 'infants', 'water', 'budget', 'physician','salvador', 'religious', 'satellite', 'aid', 'missile',
        'immigration', 'synfuels', 'education', 'superfund', 'crime', 'duty_free_exports','eaa_rsa']
        
df = pd.read_csv(url, names = names)

df.head()
```

```python
df.replace(('y', 'n', '?'), (1, 0, np.NaN), inplace=True)
df.head()
```

```python
df.fillna(method='bfill', inplace=True)
df.tail()
```

```python
df.fillna(method='ffill', inplace=True)
df.tail()
```

```python
df.party.replace(('republican', 'democrat'), (1, 0), inplace=True)
df.info()
```

```python
for col in df.columns:
  df[col] = pd.to_numeric(df[col]).astype(int)

df.info()
```

```python
y = df.party.values # labels
X = df.drop('party', axis=1) # features
```

```python
class Classifier:
  def __init__(self, X, y):
    from sklearn.model_selection import train_test_split
    self.Xtrain,self.Xtest,self.ytrain,self.ytest = train_test_split(X, y, test_size = 0.25, random_state = 42, stratify=y)
  
  def kNN(self, k = 5):
    print('KNN Algorithm')
    from sklearn.neighbors import KNeighborsClassifier
    KNN = KNeighborsClassifier(n_neighbors=k)
    KNN.fit(self.Xtrain, self.ytrain)
    self.y_pred = KNN.predict(self.Xtest)

    self.report()

  def svm(self, kernel = 'rbf'):
    print('SVM Algorithm')
    from sklearn.svm import SVC
    classifier = SVC(kernel = kernel)
    classifier.fit(self.Xtrain, self.ytrain)
    self.y_pred = classifier.predict(self.Xtest)

    self.report()

  def randomForest(self, n_estimators, criterion, random_state):
    print('Random Forest Algorithm')
    from sklearn.ensemble import RandomForestClassifier
    classifier = RandomForestClassifier(n_estimators = n_estimators, criterion = criterion, random_state = random_state)
    classifier.fit(self.Xtrain, self.ytrain)
    self.y_pred = classifier.predict(self.Xtest)

    self.report()

  def classificationReport(self):
    from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
    print("Accuracy:", accuracy_score(self.ytest, self.y_pred))
    print(classification_report(self.ytest, self.y_pred))

  def confusionMatrix(self):
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(self.ytest, self.y_pred)
    print(cm)

  def report(self):
    self.confusionMatrix()
    self.classificationReport()
```

```python
classifier = Classifier(X, y)
```

```python
classifier.kNN(k=2)
```

```python
classifier.svm()
```

```python
classifier.randomForest(n_estimators = 10, criterion = 'entropy', random_state = 0)
```
